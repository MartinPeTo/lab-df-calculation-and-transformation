## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
customer_spent = orders.groupby(['CustomerID'])[['amount_spent','Quantity']].agg('sum')
customer_spent.head()
```

```python
def quantile_spent(x):
    return customer_spent[customer_spent['amount_spent'] > customer_spent['amount_spent'].quantile(x)]['amount_spent']

quantile_spent(0.90).head()
```

```python
#Definimos los VIP por encima del p90 y preferred por encima del p75

p90 = customer_spent['amount_spent'].quantile(0.9)
p75 = customer_spent['amount_spent'].quantile(0.75)

def customer_label(x):
    if x > p90:
        return 'VIP'
    elif x > p75:
        return 'Preferred'
    else:
        return 'Cutre'

customer_spent['customer_type'] = customer_spent['amount_spent'].apply(customer_label)
customer_spent.head()
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

# Q2: How to identify which country has the most VIP Customers?

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

Provide your solution for Q2 below:

```python
# your code here
orders=orders.join(customer_spent['customer_type'],on = 'CustomerID')
orders.head()
```

```python
orders.groupby(['Country','customer_type']).agg('count')
```

```python
count_vips = orders[orders.customer_type=='VIP'].groupby(['Country'])['customer_type'].agg('count')
```

```python
count_vips.sort_values(ascending = False).head(1)
```

```python
#Q3 answer
count_preferred = orders[orders.customer_type=='Preferred'].groupby(['Country'])['customer_type'].agg('count')
```

```python
count=count_vips+count_preferred
count.sort_values(ascending = False).head(1)
```
